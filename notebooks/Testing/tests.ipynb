```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kde

# Generate some random data
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)

# Create a 2D kernel density estimator
k = kde.gaussian_kde(np.vstack([x, y]))

# Evaluate the KDE on a grid
xgrid, ygrid = np.mgrid[x.min():x.max():100j, y.min():y.max():100j]
z = k(np.vstack([xgrid.flatten(), ygrid.flatten()]))

# Reshape the result to a 2D array
z = z.reshape(xgrid.shape)

# Plot the density map
fig, ax = plt.subplots()
im = ax.imshow(z, cmap='inferno', extent=[x.min(), x.max(), y.min(), y.max()])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.colorbar(im)

plt.show()
```

```python
import numpy as np
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt

# create a 2D matrix with random values
data = np.random.randn(2, 1000)

# perform kernel density estimation
kde = gaussian_kde(data)

# create a grid of points to evaluate the KDE at
x, y = np.mgrid[-3:3:100j, -3:3:100j]
grid = np.vstack([x.ravel(), y.ravel()])

# evaluate the KDE at the grid points
z = kde(grid)

# plot the results
fig, ax = plt.subplots()
ax.imshow(z.reshape(100, 100), origin='lower', extent=[-3, 3, -3, 3])
ax.plot(data[0, :], data[1, :], 'k.', markersize=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
import seaborn as sns
from sklearn.neighbors import KernelDensity

n = 10000
dist_frac = 0.1
x1 = np.random.normal(-5,2,int(n*dist_frac))
x2 = np.random.normal(5,3,int(n*(1-dist_frac)))
x = np.concatenate((x1,x2))
np.random.shuffle(x)
eval_points = np.linspace(np.min(x), np.max(x))

kde_sk = KernelDensity(bandwidth=1, kernel='gaussian')
kde_sk.fit(x.reshape([-1,1]))
y_sk = np.exp(kde_sk.score_samples(eval_points.reshape(-1,1)))

kde_sp = gaussian_kde(x, bw_method='scott')     ### I MEAN HERE! ###
y_sp = kde_sp.pdf(eval_points)

sns.kdeplot(x)
plt.plot(eval_points, y_sk)
plt.plot(eval_points, y_sp)
plt.legend(['seaborn','scikit','scipy'])
```
